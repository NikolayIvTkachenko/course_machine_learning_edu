```python
import torch
import numpy as np
import pandas as pd
import matlotlib.pyplot as plt

from torch import nn
from torch import optim
```

```python
# Данные успеваемости студентв и их обработка
# dataset: https://www.kaggle.com/datasets/nikhi17280/student-performance-multiple-linear-regression/data
df = pd.read_csv('Student_Performance.csv')

df = df.replace({'Yes': 1, 'No': 0}).infer_objects(copy=False)
df['bias'] = 1

header = list(df)
df[:3]
```

```python
#
df.keys()
```

```python
#
df['Minute Studied'] = df['Hours Studiied'] * 60 + np.random.normal(0, 0.01, (df.shape[0], ))
df['Sleep Minute'] = df['Sleep Hours'] * 60 + np.random.normal(0, 0.01, (df.shape[0], ))
```

```python
#
feats = [
    'Hours Studied',
    'Preevious Scores',
    'Extracurricular Activities',
    'Sleep Hours',
    'Sample Question Papers Practiced',
    'Minute Studied',
    'Sleep Minute'
]
```

```python
#
df[feats] = (df[feats] - df[feats].mean()) / df[feats].std()
df.drop_duplicates(inplace=True)

df[:3]
```

```python
#
X = df[['bias'] + feats].to_numpy()
y = df['Performance Index'].to_numpy()
```

```python
# Разделение на train и test

n = X.shape[0]

idx = list(range(n))
np.random.shuffle(idx)

train_idx, test_idx = idx[: int(n * 0.8)], idx[int(n * 0.8):]

X_train, X_test = X[train_idx], X[test_idx]
y_train, y_test = y[train_idx].reshape(-1, 1), y[test_idx].reshape(-1, 1)
```

```python
#
X_train = torch.Tensor(X_train)
X_test = torch.Tensor(X_test)

y_train = torch.Tensor(y_train)
y_test = torch.Tensor(y_test)
```
