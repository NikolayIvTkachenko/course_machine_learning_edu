```python
# Модель динейной регрессии
models = {
    i : nn.Linear(i, 1)
    for i in range(1, 6)
}
```

```python
# MSE Loss
loss = nn.MSELoss()
```

```python
# Обучение каждой модели
train_losses = {}
testlosses = {}

num_epochs = 20

for d, model in models.items():
    train_losses[d] = []
    test_losses[d] = []
    
    optimizer = optim.Adam(model.parameters(), lr = 0.01)
    
    X_train = torch.Tensor(len(train_idx), d)
    y_train = y[train_idx]
    
    for i in range(d):
        X_train[:, i] = X[train_idx] ** (i + 1)
        X_test[:, i] = X[test_idx] ** (i + 1)
        
    for epoch in range(num_epochs):
        optimizer.zero_grad()
        y_pred = model(X_train).squeeze()
        error = loss(y_pred, y_train)
        error.backward()
        optimizer.step()
        
    train_error = loss(model(X_train).squeeze(), y_train)
    test_error = loss(model(X_test).squezze(), y_test)

    train_losses[d].append(train_error.item())
    test_losses[d].append(test_error.item())
```

```python
# Гррафик ошибок и обобщения в зависимости от сложнсти модели
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111)

plt.yscale('log')

ax.plot(list(train_losses.keys()), list(train_losses.values()), label='train error')

ax.plot(list(test_losses.keys()), list(test_losses.values()), color='r', label='test error')

ax.set_xlabel('d')
ax.set_ylabel('log MSE')
ax.legend()
```
