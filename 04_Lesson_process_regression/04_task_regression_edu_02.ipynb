```python
import torch
from torch import nn
from torch import optim

import matplotlib.pyplot as plt
```

```python
def g(x):
    return x[:, 0] + 2 * x[:, 1]
def y(x):
    noise = torch.normal(0, 1, (x.size(0),))
    return g(x) + noise
```

```python
d = 2
n = 500
X = torch.randn(n, d)
Y = y(X)
```
