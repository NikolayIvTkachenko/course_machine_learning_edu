```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
```

```python
# Генерация случайных двумерных данных
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
# Обучение модели линейнй регрессии
model = LinearRegression()
model.fit(X, y)
```

```python
# Предсказание на обучающих данных

y_pred = model.predict(X)
```

```python
# Вычисление метрик
mse = mean_squared_error(y, y_pred)
mae = mean_absolute_error(y, y_pred)
r2 = r2_score(y, y_pred)
```

```python
# Визуализация данных и регрессионной прямой
plt.scatter(X, y, color='blue', label='Данные')
plt.plot(X, y_pred, color='red', linewidth=3, label='Линейная регрессия')
plt.title('Линейная регрессия')

plt.xlabel('Признак X')
plt.ylabel('Целевая переменная y')
plt.legend()
plt.show()
```

```python
# Вывод метрик
print(f'Mean Squared Error (MSE): {mse:.2f}')
print(f'Mean Absolute Error (MAE): {mae:.2f}')
print(f'R^2 score: {r2:.2f}')
```
