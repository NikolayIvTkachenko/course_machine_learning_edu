```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta
```

```python
# Задаем параметры генерации данных
num_rows = 1000

# Генерация количества данных
quantitative_data = {
    'age': np.random.randint(18, 70, size=num_rows),
    'income': np.random.randint(20000, 1150000, size=num_rows),
    'height': np.random.normal(170, 10, size=num_rows), # нормальное распредлеение роста
    'weight': np.random.normal(70, 15, size=num_rows),# нормальное рраспределение веса
}
```

```python
# Генерация категориальных данных
categories = ['cat', 'dog', 'bird', 'fish']
categorical_data = {
    'pet_type': [random.choice(categories) for _ in range(num_rows)],
    'education_level': np.random.choice(['High School', 'Bachelor', 'Master', 'PhD'], size=num_rows),
    'employment_status': np.random.choice(['Employed', 'Unemployed', 'Self-employed'], size=num_rows),
}
```

```python
#  Генерация бинарных данных
binary_data = {
    'is_married': np.random.choice([0, 1], size=num_rows),
    'has_children': np.random.choice([0, 1], size=num_rows),
    'has_insurance': np.random.choice([0, 1], size=num_rows),
}
```

```python
# Генерация врременных данных (даты рождения)
start_date = datetime(1970, 1, 1)
end_date = datetime(2000, 12, 31)
time_data = {
    'birth_data': [start_date + timedelta(days=random.randint(0, (end_date - start_date).days)) for _ in range(num_rows)]
}
```

```python
# Генерация географических данных (страны и города)
countries = ['USA', 'Canada', 'UK', 'Germany', 'France']

cities = ['New York', 'Los Angeles', 'Toronto', 'London', 'Berlin']
geographical_data = {
    'country': np.random.choice(countries, size=num_rows),
    'city': np.random.choice(cities, size=num_rows),
}
```

```python
# Собираем все данные в один DataFrame
data = pd.DataFrame({**quantitative_data, **categorical_data, **binary_data, **time_data, **geographical_data})
```
